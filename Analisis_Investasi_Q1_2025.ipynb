## 📌 Pendahuluan
Memahami data realisasi investasi bukan hanya soal angka triliunan Rupiah, tetapi soal memetakan **peluang dan risiko**. Bagi sebuah bank, daerah dengan realisasi investasi tinggi adalah 'kolam' nasabah yang kaya akan likuiditas. Sebaliknya, sektor yang stagnan memberikan sinyal risiko gagal bayar yang lebih tinggi.

Notebook ini memproses data Triwulan I 2025 untuk menjawab pertanyaan:
1. **Di mana** kita harus membuka cabang baru atau memperkuat tim sales?
2. **Siapa** target nasabah korporat yang paling sehat saat ini?
3. **Industri apa** yang memiliki penyerapan tenaga kerja tinggi (potensi *Payroll* & *Retail Lending*)?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Konfigurasi Visualisasi Premium
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.titlepad'] = 20
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
```

## 1. Integrasi & Pembersihan Data
Kita memuat data dan melakukan standarisasi pada kolom-kolom kunci agar siap dianalisis.

```python
file_path = 'Data Realisasi Investasi Triwulan I Tahun 2025.csv'
df = pd.read_csv(file_path)

# Pembersihan Tipe Data
df['investasi_rp_juta'] = pd.to_numeric(df['investasi_rp_juta'], errors='coerce').fillna(0)
df['tki'] = pd.to_numeric(df['tki'], errors='coerce').fillna(0)

print(f"Total Baris Data: {len(df)}")
df.head()
```

## 2. Peta Kekuatan Ekonomi Regional
Daerah dengan aliran modal tinggi biasanya memiliki pertumbuhan ekonomi yang stabil.

```python
top_prov = df.groupby('provinsi')['investasi_rp_juta'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
colors = sns.color_palette("Blues_r", len(top_prov))
sns.barplot(x=top_prov.values, y=top_prov.index, palette=colors)
plt.title('🎯 Top 10 Provinsi: Prioritas Alokasi Sumber Daya Bank')
plt.xlabel('Nilai Investasi (Juta Rp)')
plt.show()
```

### Karakteristik Jawa vs Luar Jawa
- **Jawa:** Biasanya didominasi oleh Industri Pengolahan & Jasa (Tersier).
- **Luar Jawa:** Biasanya didominasi oleh Sumber Daya Alam (Primer/Sekunder).

```python
island_dist = df.groupby('pulau')['investasi_rp_juta'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
island_dist.plot(kind='bar', color='teal')
plt.title('Distribusi Investasi Per Pulau')
plt.ylabel('Nilai Investasi (Juta Rp)')
plt.xticks(rotation=45)
plt.show()
```

## 3. Analisis Sektor: Fokus Industri
Menentukan sektor mana yang menjadi 'pemain utama' di Q1 2025.

```python
top_sector = df.groupby('nama_sektor')['investasi_rp_juta'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(12, 10))
sns.barplot(x=top_sector.values, y=top_sector.index, palette='viridis')
plt.title('🚀 15 Sektor Industri Terpanas (Target Nasabah Korporat)')
plt.xlabel('Nilai Investasi (Juta Rp)')
plt.show()
```

## 4. Potensi Nasabah Retail & Payroll
Sektor yang menyerap banyak Tenaga Kerja Indonesia (TKI) adalah target utama untuk produk pinjaman konsumsi, KPR, dan kartu kredit.

```python
tki_sector = df.groupby('nama_sektor')['tki'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=tki_sector.values, y=tki_sector.index, palette='Oranges_r')
plt.title('👥 Sektor Padat Karya: Peluang Kerjasama Payroll')
plt.xlabel('Jumlah Tenaga Kerja (Orang)')
plt.show()
```

## 5. Alat Bantu Analisis Cabang (Interactive Analysis)
Fungsi di bawah ini dapat digunakan oleh Relationship Manager untuk menganalisis daerah cakupan cabang mereka secara spesifik.

```python
def analyze_branch_area(city_name):
    # Filter data berdasarkan kecocokan nama kota
    city_df = df[df['kabupaten_kota'].str.contains(city_name, case=False, na=False)]
    
    if city_df.empty:
        return f"Data untuk {city_name} tidak ditemukan."
    
    top_inv = city_df.groupby('nama_sektor')['investasi_rp_juta'].sum().sort_values(ascending=False)
    top_lab = city_df.groupby('nama_sektor')['tki'].sum().sort_values(ascending=False)
    
    print(f"--- ANALISIS CABANG: {city_name.upper()} ---")
    print(f"Total Investasi di Area: Rp {city_df['investasi_rp_juta'].sum():,.0f} Juta")
    print(f"🥇 Sektor Unggulan: {top_inv.index[0]} (Rp {top_inv.iloc[0]:,.0f} Juta)")
    print(f"👷 Sektor Padat Karya: {top_lab.index[0]} ({top_lab.iloc[0]:,.0f} Pekerja)")
    
    # Visualisasi mini
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    sns.barplot(x=top_inv.head(5).values, y=top_inv.head(5).index, ax=ax[0], palette='crest')
    ax[0].set_title('Top 5 Sektor (Investasi)')
    
    sns.barplot(x=top_lab.head(5).values, y=top_lab.head(5).index, ax=ax[1], palette='flare')
    ax[1].set_title('Top 5 Sektor (Penyerapan Kerja)')
    
    plt.tight_layout()
    plt.show()

# CONTOH PENGGUNAAN:
analyze_branch_area('Bekasi')
```

## 6. Ringkasan Strategis & Mitigasi Risiko

### 🟢 High Potential (Target Kejar)
Cari perusahaan yang bergerak di sektor di bawah ini pada daerah dengan investasi tertinggi:
1. **Industri Logam Dasar & Elektronik**: Biasanya memiliki capex besar (Peluang Kredit Investasi).
2. **Pertambangan & Perkebunan (Luar Jawa)**: Membutuhkan *Trade Finance* & *Cash Management*.
3. **Jasa & Transportasi (Jawa)**: Membutuhkan solusi digital & *Payroll*.

### 🔴 Red Flags (Hati-hati Akuisisi)
1. **Sektor Stagnan**: Sektor yang tercatat dalam data tetapi memiliki nilai investasi Rp 0 atau sangat kecil di Q1. Ini indikasi industri tersebut sedang mengerem ekspansi.
2. **Daerah dengan Low Absorption**: Daerah dengan investasi besar tapi nol penyerapan tenaga kerjanya (padat modal ekstrim), mungkin kurang menguntungkan untuk nasabah individual/retail.

### 💡 Action Plan untuk Branch Manager:
- **Mapping Komunitas**: Identifikasi asosiasi pengusaha di sektor unggulan daerah Anda.
- **Cross-Selling**: Gunakan data TKI untuk masuk ke HRD perusahaan tersebut guna menawarkan *Payroll bundle*.
- **Monitoring**: Jika ada penurunan investasi drastis di sektor andalan daerah Anda, segera tinjau ulang profil risiko debitur terkait.
